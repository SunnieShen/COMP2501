We may be contempt with the state of the data at this point. However, if we were to run a statistical analysis now, we would realise that things are not quite like we would like them to be. These object/string variables are in fact categorical or ordinal. If we do not specify the order of their categories/levels, statistical packages may choose the wrong reference category or order. Thus, it will be benefitial to convert these variables to the right type and set the desired order of their categories.

This can be done manually for each variable, or we can identify and exploit patterns. This becomes particularly useful with increased number of variables. Looking at the variables we identify that:

1. `'work_interfere'`, `'no_employees'` and `'leave'` are [ordinal](https://duckduckgo.com/?q=R+ordinal+variable&t=newext&atb=v330-1&ia=webhttps://duckduckgo.com/?q=R+ordinal+variable&t=newext&atb=v330-1&ia=web). These need to be set manually.

2. `'Country'` and `'Gender'` are [categorical](https://duckduckgo.com/?q=R+categorical+variable&t=newext&atb=v330-1&ia=webhttps://duckduckgo.com/?q=python+pandas+categorical+variable&t=newext&atb=v330-1&ia=web), and the order of their categories are not important. However, it will be interesting to have the largest level as the reference when doing statistical modelling. Usually, the first level is used as reference.

3. When the variable has 'Yes' and 'No' levels, the variable is categorical. Level 'No' should be the reference. An optinal third level (often denoting uncertainty) will be located in the middle.

In terms of coding, we will:

1. Loop through all the columns.
2. Focus on the ones that are of type string.
3. Check column names to identify the cases enumerated above and take the right action.

```r
# loop through all column names
for (c in colnames(csv_data)){
    
    # Check if the selected column is character
    if (is.character(csv_data[, c])){
           
        # Read unique values (i.e., categories)
        cats = unique(csv_data[, c])
        
        # TIP: Order conditionals from more to less specific. Why?
        # work_interfere is ordinal
        if (c == 'work_interfere'){
            
            # Convert to ordinal with order: 'Never', 'Rarely', 'Sometimes', 'Often'
            csv_data[c] = factor(csv_data[, c], order=TRUE, levels=c('Never', 'Rarely', 'Sometimes', 'Often'))
        
        # no_employees is ordinal    
        } else if (c == 'no_employees') {
            
            # Convert to ordinal with order: '1-5', '6-25', '26-100', '100-500', '500-1000', 'More than 1000'
            csv_data[c] = factor(csv_data[, c], order=TRUE, levels=c('1-5', '6-25', '26-100', '100-500', '500-1000', 'More than 1000'))
        
        # leave is ordinal
        } else if (c == 'leave') {
            
            # Convert to ordinal with order: 'Very difficult', 'Somewhat difficult', "Don't know", 'Somewhat easy', 'Very easy'
            csv_data[c] = factor(csv_data[, c], order=TRUE, levels=c('Very difficult', 'Somewhat difficult', "Don't know", 'Somewhat easy', 'Very easy'))
        
        # Country is categorical, but it will be good to use the biggest category as reference. Thus ordering the categories
        # as 'United States', 'United Kingdom', 'Canada', and 'Other'
        } else if (c == 'Country') {
            
            # Convert to categorical with order: 'United States', 'United Kingdom', 'Canada', 'Other'
            csv_data[c] = factor(csv_data[, c], order=FALSE, levels=c('United States', 'United Kingdom', 'Canada', 'Other'))
                    
        # Gender is categorical, but it will be good to use the biggest category as reference (i.e. ordered as 'Male', 'Female').
        } else if (c == 'Gender') {
            
            # Convert to categorical with order 'Male', 'Female'
            csv_data[c] = factor(csv_data[, c], order=FALSE, levels=c('Male', 'Female'))
                        
        # If the variable has 'Yes' and 'No' categories, then we will convert it 
        # to a categorical variable with levels 'No', CAT3, 'Yes', where CAT3 is
        # whatever third category the variable may have        
        } else if (('Yes' %in% cats) & ('No' %in% cats)){
            
            # Retrieve the thrid category
            cat3 = cats[(cats != 'No') & (cats != 'Yes')]
            # Check that a third category actually exist. If cat3 is empty, the it doesn't
            if (length(cat3)==0){
                levels = c('No', 'Yes')
            } else {
                levels = c('No', cat3, 'Yes')
            }
            # Convert to categorical/factor
            csv_data[c] = factor(csv_data[, c], levels=levels)
            
        # The remaining variables can have any order
        } else {
            
            # Convert to categorical/factor
            csv_data[c] = factor(csv_data[, c])
            
        }
        
    }
    
}
```

Finally, let us move `'treatment'` to the front of the table as we will use this as the dependent variable in our statistical analysis.

```r
# Get list of all columns
cols = colnames(csv_data)
# Build a new list of columns in which 'treatment' appears first
reordered_cols = c(c('treatment'), cols[cols != 'treatment'])
# Re-order columns in the data.frame
csv_data = csv_data[, reordered_cols]

# Let us have a look at the summary again
print(summary(csv_data))
```

## Functions: actions repeated across scripts

We want to replicate the previous plots for all numerical and categorical variables. We could do this manually with *a lot* of copy&paste **OR** we could do this programmatically, i.e. the right way! :-) Doing this programatically means:

1. Loop through all the columns in the `DataFrame`.
2. For each column/variable, check the data type and plot the correct figure type.
3. Rather than having many figures, we would like to have all the plots in a single [figure with one panel per data variable/column](https://duckduckgo.com/?q=R+ggplot2+multiple+panels+in+figure&t=newext&atb=v330-1&iar=images&iax=images&ia=images).

Let us assume that this is actually something we would like to do for all projects. In such case, we would like to make this into a [function](https://duckduckgo.com/?q=R+function&t=h_&ia=web) that can then be called with any`DataFrame`. The function will need to know the name of the independent variable, since this variable will be used to group the data in the figures (as above). The function will also need to work out how many panels will be plotted, how these will be distributed in the figure (panel grid) and how big the figure should be to ensure panels are large enough to be read. The following code snippet on its own represent a script that can be saved in a separate file (e.g. mytoolbox.R) which can then be imported in future scripts/projects. Make sure you also include [documentation](https://duckduckgo.com/?q=R+add+function+documentation&t=h_&ia=web).

```r
# Import all necessary libraries
library(ggplot2)
# library("gridExtra")
library(cowplot)


#' Plot distribution of all variables in the data grouping by an independent categorical variable.
#'     
#'  Parameters
#'  ----------
#'      data : DataFrame
#'          data table to plot
#'      dv : str
#'          name of dependent categorical variable in `data`
plot_data <- function(data, dv){

    # Check `data` us if type `data.frame`
    if (!is.data.frame(data)){
        stop(paste("`data` must be of type `data.frame`. The passed variable is", class(data), "instead."))
    }
    # Check `dv` is of type `str`
    if (!is.character(dv)){
        stop(paste("`dv` must be of type `character`. The passed variable is", class(dv), "instead."))
    }
    # Check `dv` is an existing column in `data`
    if (!(dv %in% colnames(data))){
        stop("The specified `dv` is not in the `data`.")
    }

    # Get list of columns to plot
    col_types = sapply(data, class)
    # Check that the dependent variable is categorical
    if (col_types[[dv]] != 'factor'){
        # Raise an error message to help the user understand what's wrong
        stop(paste("The independent variable must be of type factor. The specified column is", col_types[[dv]], "instead."))
    }
    # Remove independent variable from `col_types`, as this will not be separately plotted
    col_types[[dv]] <- NULL

    # Calculate number of rows and columns in the figure
    n = length(col_types)
    nrows = floor(sqrt(n))  # Floor rounding
    ncols = ceiling(n/nrows)  # Ceil rounding
    
    # Allocate a vector to place the results
    plots = vector(length=length(col_types), "list")

    # Plot each columns
    coln = 0
    for (col_name in names(col_types)){  # col_name = names(col_types)[[2]]
        coln = coln + 1
        # Check if the current column (col_n and col_name) is of numeric type
        if (col_types[col_name] == 'numeric'){
            p = ggplot(data, aes_string(x=col_name, color=dv, fill=dv)) +
              geom_histogram(alpha=.5, position="identity", binwidth=2)
        # Check if the current column (col_n and col_name) is categorical
        } else if (any(col_types[[col_name]] == 'factor')){
            count_data = data %>% group_by_at(c(dv, col_name)) %>% count()
            p = ggplot(count_data, aes_string(x = dv, y = 'n', color = col_name, fill = col_name)) + 
                geom_bar(position = "fill",stat = "identity")
        }
        # Append plot
        plots[[coln]] = p
    }

    # Plot grid
    args = append(plots, list(labels=names(col_types), ncol=ncols, nrow=nrows))
    # labels=labels
    do.call(plot_grid, args)
}
```

This can now be used anywhere in your code.

```r
# Ignore the following line. This is just to change the size of the figure in Kaggle.
options(repr.plot.width = 20, repr.plot.height = 15)

# Use our function
plot_data(csv_data, 'treatment')
```

The power of having this as a function, is that the code is agnostic to the data and the selected dependent variable, as long as they meet the criteria specified in the function's docstring. Specifically, we will build a [logistic regression model](https://duckduckgo.com/?q=R+stats+logistic+regression&t=newext&atb=v330-1&ia=web) using `'treatment'` as the dependent variable, and all the other variables as independent variables. Let us start with the empty model for simplicity.

```r
# Build the logistic regression model
log_reg = glm(treatment ~ 1, csv_data, family=binomial)
```

Let us print a [summary of the model](https://duckduckgo.com/?q=R+stats+print+model+summary&t=newext&atb=v330-1&ia=web).

```r
# Print a summary of the model
print(summary(log_reg))
```

Next, we need to add all the variables to the equation. We can do this manually, or we can add them programatically. For the latter, we need to build a text strig of the form `dv` = `iv1` + `iv2` + `iv3` + `...`, where `dv` is the name of the dependent variable (in this case 'treatment') and `ivn` are the names of all the other columns in the data.

```r
# Retrieve a list with all columns except 'treatment'
cols = colnames(csv_data)
cols = cols[cols != 'treatment']

# Build string that will serve as the model equation: dv = iv1 + iv2 + iv3 + ... with dv = 'treatment' 
# (as specified in the snippet above) and iv as all the independent variables
eq = paste("treatment ~", paste(cols, collapse = ' + '))

# Build the logistic regression model
log_reg = glm(eq, csv_data, family=binomial)

# Print a summary of the model
print(summary(log_reg))
```
